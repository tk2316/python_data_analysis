```python
# np 정보 출력용으로 사용할 함수를 정의한다.
def pprint(arr):
    print('type: {}'.format(type(arr)))
    print('shape: {}, dimension: {}, dtype: {}'.format(arr.shape, arr.ndim, arr.dtype))
    print('numpy 배열에 저장된 데이터\n', arr)
```

파이썬 list 객체로 numpy 배열을 생성할 수 있다.
인수로 list 객체와 데이터 타입(dtype)을 입력해 numpy 배열을 생성하고 dtype을 생략할 경우 입력되는 list 객체의 타입으로 설정된다.

```python
#파이썬 1차원 list로 numpy 배열 만들기
arr = [1,2,3]
print(type(arr))

# np.array() 함수는 인수로 list객체를 넘겨받아 numpy배열을 생성한다.
a = np.array(arr)
#dtype안쓰고 숫자를 입력하면 기본값 int32
#a = np. array(arr, dtype=int)
#a = np. array(arr, dtype=np.int32)
#이렇게 써줘도 문제없다
pprint(a)

```

```python
#파이썬 2차원 list로 numpy 배열 만들기
#arr = [["1."]] 이렇게 적으면 dtype이 기본으로 float64로 나온다
#정확하게 dtype를 int로 (예를들면) 지정하면 다 인트로 바뀐다.

arr = [[1.,2,3],[4,5,6]]
print(type(arr))

a = np.array(arr)
pprint(a)

```

```python
#파이썬 3차원 list로 numpy 배열 만들기
arr = [[[1,2,3], [4,5,6]],[[11,22,33],[44,55,66]]]
print((type(arr)))
a = np.array(arr, dtype =np.int16)
print(type(a))
pprint(a)
```

numpy는 원하는 shape으로 배열을 생성하고 각 요소를 특정값으로 초기화하는 함수를 제공한다.


```python
#np.zeros() 함수는 지정된 shape의 배열을 생성하고 모든 요소를 0으로 초기화한다
#np.zeros(shape, dtype, order)
#order => 다차원 데이터를 행 운선 방식(c/c++, java, python), 열 우선 방식(fortran)으로 메모리에 저장할지 여부를 결정한다.
#a = np.zeros(shape=(3,4), dtype=np.int16, order='f')
#3행 4열
a = np.zeros(shape=(3,4), dtype=np.int32)
pprint(a)
```

```python
a = np.ones(shape=(3,4), dtype=np.int32)
pprint(a)
```

```python
#np.full()함수는 지정된 shape의 배열을 생성하고 모든 요소를 특정값으로 초기화시킨다
#a = np.full(shape, fill_value, dtype, order)
a = np.full(shape=(3,4),fill_value=999, dtype=np.int32)
pprint(a)
```

```python
#np.eye()함수는 shape이(n,n)인 단위행렬(대각선 요소만 1이고 나머지는 0인 행렬)을 생성한다
#np.eye(n)
a = np.eye(5)
pprint(a)
```

```python
#np.empty()함수는 지정된 shape의 배열을 만든다 => 초기화를 하지 않는다 => 쓰레기가 저장되어 있다
#np.empty(shape, dtype, order)
a = np.empty(shape=(3, 4))
pprint(a)
```

```python
#like 함수
#numpy는 지정된 배열과 shape이 같은 행렬을 만드는 *_like 함수를 제공한다.
#np.zeros_like(), np.ones_like(), np.full_like(), np.empty_like()
a = np.array([[1,2,3],[4,5,6]])
pprint(a)
b = np.zeros_like(a)
pprint(b)

c = np.ones_like(a)
pprint(c)

d = np.full_like(a, 777)
pprint(d)

e = np.empty_like(a)
pprint(e)
```

데이터 생성 함수

```python
#numpy는 주어진 조건으로 데이터를 생성한 후 배열을 만드는 데이터 생성 함수를 제공한다.
#np.linspace, np.arange, np.logspace
#!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
#np.linspace() 함수는 start부터 stop의 범위에서 num개를 균일한 간격으로 데이터를 생성해서 배열을 만든다
#np.linspace(start,stop,num,endpoint = True, retstep= True)
#endpoint는 기본으로 True이며 끝을 'end'포함한다 (5포함)
#retstep은 간격을 보여준다  => (array([0.  , 0.25, 0.5 , 0.75, 1.  ]), 0.25) 좌측처럼 결과를 보여주는데 문제는 튜플형식이다. 그래서 plot이 안됀다

#num을 생략하면 기본 값은 50이다
#요소의 개수를 기준으로 균등 간격 배열을 생성한다
a = np.linspace(0, 1, 5)   #0부터 1까지 5등분해라
print(a)
print(type(a))
#np.linspace() 함수 실행결과 시각화
plt.plot(a, '*')    #문자열로 색갈 선택, "d"는 다이아몬드, "h"하면 동그라미, "*"하면 별모향
plt.show()

#다른 parameter 포함한 예시
a = np.linspace(0, 1, 5, endpoint=True, retstep = True)   #0부터 1까지 5등분해라
print(a)
print(type(a))
#여기서 numpy가 아닌 tuple type이다
#0.24땜에 히스토그램을 그리지 못한다 그러하며 이것의 해결책은 아래있다
plt.plot(a[0], 'g*')
plt.show()
```

```python
#np.arange() 함수는 start부터 stop 미만까지 step 간격으로 데이터를 생성하고 배열을 만든다
#np.arange([start,] stop [,step], dtpye)   start, step, dtype생략가능
#np.linspace() 함수는 요소의 개수로 배열을 만들었지만 np.arange()함수는 데이터 간격을 기준으로 균동한 간격의 배열을 생성한다.
#start를 생략하면 0이 기본값, step도 생략하면 1이 기본값.
#np.arange(10), np.arange(0,10), np.arange(0,10,1)은 모두 같은 기능이 실행된다
a = np.linspace(0,10,5)
pprint(a)

b = np.arange(0,10,2.5)
pprint(b)
plt.plot(b, "g*")
plt.show()
```

```python
#np.logspace() 함수는 로그 스케일로 지정된 범위(start 부터 stop사이)에서 num개 만큼 균등 간격으로 데이터를 생성하고 배열을 
#만든다.
#np.logspace(start, end, num, base)
a = np.logspace(1,10, num=10, base=2)
pprint(a)
plt.plot(a)
```

난수 (random number)기반 배열 생성 
creating arrangements using random numbers (난수)

```python
#numpy는 난수 (랜던) 발생 및 배열을 생성하는 random 모듈을 제공하고 다음과 같은 함수를 사용할 수 있다
#np.random.normal(), np.random.randn(), np.random.rand(), np.random.randint(), np.random.random()
```

```python
#np.random.normal() 함수는 정규 분포를 (normal distribution) 따르는 난수로 데이터를 생성하고 배열을 만든다
#np.random.normal(loc, scale, size)      => size는 count 총 합을 말한다
#loc => 정규 분포의 평균 (mu), scale => 표준편차 (standard deviation), size => 행렬의 크기 (난수의 개수)
a = np.random.normal(0, 1, 10)
pprint(a)
plt.hist(a)
plt.show()

#np.random.normal() 함수가 생성한 난수는 정규 분포의 항상을 가진다
help(np.random.normal)
```

```python
#np.random.normal() 함수가 생성한 난수는 정규 분포의 항상을 가진다
a = np.random.normal(0, 1, 10000)
pprint(a)
# 정규 분포에서 10000개의 표본을 뽑은 경과를 100개의 구간(bins) 구분한 히스토그램으로 표현한다
plt.hist(a, bins=2)
plt.show()
# this doesn't look like a normal distribution curve because it groups such a wide range into two bins. It does not tell us 
# a lot about the data. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin. 
# more the bin more accurately the data is depicted
#let's see from the diagram below

a = np.random.normal(0, 1, 10000)
#a stores 10000 data with mean of 0 and Std deviation of 1 that follows normal distribution model.
pprint(a)
plt.hist(a, bins = 100)
plt.show()
#by adding more bins we decrease count scale so we can see more the counts in more detail.
```

```python
#np.random.randn()함수는 표준 정규 분포(평균이 0이고 표준편파가 1)를 따르는 난수로 데이터를 생성하고 배열을 만든다
#np.random.randn(d0, d1, d2, .., dn)
##일차원도 만들수있다
a = np.random.randn(10)
##이차원도 만들수있다
#a = np.random.randn(2,3)
##3차원도 만들수있는데 터짐
pprint(a)
plt.hist(a)
plt.show()
```

```python
#np.random.randn() 함수가 생성한 난수는 표준 정규 분포의 형상을 가진다
a = np.random.randn(10000)     #평균, 표준편차를 지정 안해도 됀다. 왜냐하면 표준 정규 분포의  따르는 함수입니다. 평균 0에 맞쳐있다 
pprint(a)
# 정규 분포에서 10000개의 표본을 뽑은 경과를 100개의 구간(bins) 구분한 히스토그램으로 표현한다
plt.hist(a, bins=100)
plt.show()
```

```python
#np.random.rand()이 함수는 0 ~ 1 사이의 균등 분포를 (uniform distribution 사각형 비슷한거있자나) 따르는 난수로 데이터를 생성하고 배열을 만든다.
#np.random.rand(d0, d1, d2, .., dn)
#1차원
#np.random.rand() 함수가 생성한 난수는 균등 분포의 형상을 가진다
#it provides numpy array of 1 dimension that contains seemingly uniformally distributed random data that are values between 0 and 1
a = np.random.rand(10000)
pprint(a)
plt.hist(a, bins = 100)
plt.show()

#2차원도 가능한데 그래프가 이상해진다
b = np.random.rand(3,2)
pprint(b)
plt.hist(b, bins = 10)
plt.show()

```

```python
#np.random.randint() 함수는 지정된 크기만큼 low 부터 high 미만의 범위에서 정수로 난수 데이터를 생성하고 배열을 만든다.
#np.random.randint(low, high, size)

a = np.random.randint(1,46,6)
#provides a 1D numpy array of 6 random integers that are between 1 and 46. 
pprint(a)


#-100에서 100의 범위에서 난수를 발생시킨다
#provides a 1D numpy array of 10000 random integers that are between -100 and 100. 
b = np.random.randint(-100,100,10000)
pprint(b)
plt.hist(b,bins=10)
plt.show()
```

```python
#np.random.random() 함수는 0 ~ 1사이의 균등 본포에ㅔ서 난수로 데이터를 생성하고 배열을 만든다.
#np.random.random(size)
#(()) 해야한다
a = np.random.random((2,4))
pprint(a)
```

배열의 상태검사


```python
# 배열 속성 검사 항목   배열 속성 확인 방법          예시                 결과
# 배열 shape            np.ndarray.shape 속성        arr.shape            (5, 2, 3)
# 배열 길이             일차원 배열 길이 확인        len(arr)             5
# 배열 차원             np.ndarray.ndim 속성         arr.ndim             3
# 배열 요소 수          np.ndarray.size 속성         arr.size             30
# 배열 타입             np.ndarray.dtype 속성        arr.dtype            dtype('float64')
# 배열 타입 명          np.ndarray.dtype.name 속성   arr.dtype.name       float64
# 배열 타입 변환        np.ndarray.astype 속성       arr.astype(np.int)   배열 타입 변환
```

```python
#데모 배열
arr = np.random.random((2,3,6))
pprint(arr)
```

```python
# numpy 배열 객체는 다음과 같은 방법으로 속성을 확인할 수 있다
print('배열의 타입: {}'.format(type(arr)))
print('배열의 shape: {}'.format(arr.shape))
print('배열의 길이(면): {}'.format(len(arr)))          #첫번째는 면
print('배열의 길이(행): {}'.format(len(arr[0])))       #0면 => 다음은 행
print('배열의 길이(열): {}'.format(len(arr[0][0])))    #0면, 0행 => 다음은 열
print('배열의 차원: {}'.format(arr.ndim))
print('배열을 구성하는 요소의 개수: {}'.format(arr.size))
print('배열의 저장된 데이터 타입: {}'.format(arr.dtype))
#배열의 요소를 정수(int)로 변환한다 => 배열 요소의 실제 값이 변환되는 것이 아니고 화면에 보여주기만 하는 것이다
print(arr.astype(np.int))
print(arr)
#실제 데이터가 변경되게 하려면 정수로 변화한 결과를 배열에 저장시키면 된다
arr = arr.astype(np.int)
```
