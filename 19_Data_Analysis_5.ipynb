```python
missing_df_row = df[df['집행금액'].isnull()].index[0]
missing_df_row
```

```python
# 해당 행을 제거하고 다시 df 데이터프레임으로 저장한다
print(df.shape)
df = df.drop(missing_df_row)
print(df.shape)
```

```python
df
```

```python
# 집행일시 열을 이용해서 '연', '월', '연월', '일', '시', '분', '요일' 파생 변수를 만든다.
# 판다스해서 to_datatime() 함수를 사용하면 문자열 형태의 데이터를 날짜/시간 형태의 데이터로 변환할 수 있다.
df['집행일시'] = pd.to_datetime(df['집행일시'])
df['연'] = df['집행일시'].dt.year.astype(int)
df['월'] = df['집행일시'].dt.month.astype(int)
df['연월'] = df['연'].astype(str) + '-' + df['월'].astype(str)
df['일'] = df['집행일시'].dt.day.astype(int)
df['시'] =  df['집행일시'].dt.hour.astype(int)
df['분'] =  df['집행일시'].dt.minute.astype(int)
df['요일'] = df['집행일시'].dt.dayofweek.astype(int) # 0 = 월,..., 6 = 일
```

집행목적 워드클라우드

```python
from wordcloud import WordCloud
```

```python
df.head()
```

```python
#집행목적이 NaN인 데이터를 확인한다.
df[df['집행목적'].isnull()]
```

```python
# 집행목적이 NaN인 데이터가 존재하므로 NaN 공백으로 채워준다.
df['집행목적'] = df['집행목적'].fillna('')
df[df['집행목적'].isnull()]
```

```python
# 파이썬 문자열 함수 중 특정 문자열을 문자열 사이에 삽입하는 join() 함수로 집행목적을 연결해서 원드클라우드를 실행한다.
spend_gubun = ' '.join(df['집행목적'])
spend_gubun[:100]
```

```python
wordcloud = WordCloud(
    font_path = "C:\\WINDOWS\\Fonts\\NanumGothicCoding-Bold.ttf",
    background_color = 'white',
    relative_scaling = 0.5
).generate(spend_gubun)
plt.figure(figsize= (16,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
#집행장소 NaN인 데이터를 확인한다
df[df['집행장소'].isnull()]
```

```python
# NaN을 공백으로 채운다
df['집행장소'] = df['집행장소'].fillna('')
df[df['집행장소'].isnull()]
```

```python
df
```

```python
"뼈큰감자탕 (대구 중구 명덕로)".split('(')[0].strip()
```

```python
# 집행장소에 주소가 많아 보이기때문에 주소는 전처리를 통해 제거하고 '상호명'이라는 열을 만들어 집행장소에서 상호명을 가져온다.
# 상호명이 여는 괄호 앞쪽에 있으므로 여는 괄호 앞에 있는 것을 상호명으로 한다.
df['상호명'] = df['집행장소'].apply(lambda x: x.split('(')[0].strip())
df.head()
```

```python
df['상호명'].value_counts()
```

```python
# 상호명이 공백인 데이터의 집행목적을 확인한다
df.loc[df['상호명'] == '', '집행목적'].value_counts().head()
```

```python
# 상호명이 '대상자 소속부서'인 데이터의 집행목적을 확인한다.
df.loc[df['상호명'] == '대상자 소속부서', '집행목적'].value_counts().head()
```

```python
# 상호명 '-'인 데이터의 집행목적을 확인한다.
df.loc[df['상호명'] == '-', '집행목적'].value_counts().head()
```

```python
# 상호명이 공백, '대상자 소속부서', '-'인 데이터(잘못된 데이터)를 제거한다.
df_filtered = df[(df['상호명'] != '') & (df['상호명'] != '대상자 소속부서') & (df['상호명'] != '-')]
df_filtered.head()
```

```python
df_filtered['상호명'].value_counts().head(30)
```

```python
df_filtered.loc[df_filtered['상호명'] == '（주）바닷가작은부엌', '상호명'] = '바닷가작은부엌'
df_filtered['상호명'].value_counts().head(30)
```

```python
top_100 = df_filtered['상호명'].value_counts().head(100)
top_100 = pd.DataFrame(top_100).reset_index()
top_100.columns = ['상호명', '방문횟수']
top_100
```

```python
place_gubun = ' '.join(df_filtered['상호명'])
place_gubun[:100]

wordcloud = WordCloud(
    font_path = "C:\\WINDOWS\\Fonts\\NanumGothicCoding-Bold.ttf",
    background_color = 'white',
    relative_scaling = 0.5
).generate(place_gubun)
plt.figure(figsize= (16,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
