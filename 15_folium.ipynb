```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
# pip install folium
import folium
# 보고자하는 위치의 좌표계를 얻어내는 가장 단순한 방법은 구글맵을 이용하는 것이다.
```

```python
#folium 라이브러리 Map() 함수의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정한 배율을 사용하는
#지도를 만든다
#map_folium = folium.Map(location=[위도, 경도], zoom_start=배율, tiles='지도스타일')
map_folium = folium.Map(location=[37.56996025172925, 126.983705969063], zoom_start=17)

#savev() 함수로 작성된 지도를 html 파일로 html 파일로 지정할 수 있다
map_folium.save('./Output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56996025172925, 126.983705969063], zoom_start=13)

#folium 라이브러리의 Marker() 함수의 location 속성으로 지정한 위치에 마커를 만든다.
#popup 속성으로 마커를 클릭했을때 표시할 문자열을 지정할수있다
#작성한 마커는 add_to() 함수로 지도에 추가한다.
folium.Marker(location=[37.56996025172925, 126.983705969063], popup='보신각').add_to(map_folium)
folium.Marker(location=[37.56649077602773, 126.97798801324303], popup='서울시청').add_to(map_folium)
folium.Marker(location=[37.565983468856224, 126.97517828440695], popup='덕수궁').add_to(map_folium)
folium.Marker(location=[37.579804038387785, 126.97697662488285], popup='경북궁').add_to(map_folium)
folium.Marker(location=[37.577010787445786, 126.97690265925215], popup='광화문').add_to(map_folium)
folium.Marker(location=[37.57072227958591, 126.98225914764045], popup='영풍문고').add_to(map_folium)
map_folium.save('./Output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56996025172925, 126.983705969063], zoom_start=13)

# 마커를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으다면 folium 라이브러리의 Popup() 함수의 max_width 속성으로
# 변경하면 된다.
popup = folium.Popup('보신각', max_width = 200)
folium.Marker(location=[37.56996025172925, 126.983705969063], popup=popup).add_to(map_folium)

map_folium.save('./Output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56996025172925, 126.983705969063], zoom_start=13)

# icon 속성에 folium 라이브러리의 icon() 함수로 마커의 색상과 아이콘 모양을 변경할 수 있다.
# Icon() 함수의 icon 속성중 아이콘 형태도 변경할 수 있다
folium.Marker(location=[37.56996025172925, 126.983705969063], popup = folium.Popup('보신각', max_width = 200), icon = folium.Icon(color = 'red', icon ='star')).add_to(map_folium)
folium.Marker(location=[37.56649077602773, 126.97798801324303], popup=folium.Popup('서울시청', max_width = 200), icon = folium.Icon(color = 'pink', icon = 'hand-down')).add_to(map_folium)
folium.Marker(location=[37.565983468856224, 126.97517828440695], popup=folium.Popup('덕수궁', max_width = 200), icon = folium.Icon(color = 'green', icon = 'hand-up')).add_to(map_folium)
folium.Marker(location=[37.579804038387785, 126.97697662488285], popup=folium.Popup('경북궁', max_width = 200), icon = folium.Icon(color = 'blue')).add_to(map_folium)
folium.Marker(location=[37.577010787445786, 126.97690265925215], popup=folium.Popup('광화문', max_width = 200), icon = folium.Icon(color = 'black')).add_to(map_folium)
folium.Marker(location=[37.57072227958591, 126.98225914764045], popup=folium.Popup('영풍문고', max_width = 200), icon = folium.Icon(color = 'white')).add_to(map_folium)

map_folium.save('./Output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56996025172925, 126.983705969063], zoom_start=13)

# folium 라이브러리의 CircleMarker() 함수로 원 모양의 마커를 만들 수 있다
folium.CircleMarker(location=[37.56996025172925, 126.983705969063], popup = folium.Popup('보신각', max_width = 200), icon = folium.Icon(color = 'red', icon ='star'), radius = 100, colo_color = 'yellow', fill_color = 'orange').add_to(map_folium)
folium.Marker(location=[37.56649077602773, 126.97798801324303], popup=folium.Popup('서울시청', max_width = 200), icon = folium.Icon(color = 'pink', icon = 'hand-down')).add_to(map_folium)
folium.Marker(location=[37.565983468856224, 126.97517828440695], popup=folium.Popup('덕수궁', max_width = 200), icon = folium.Icon(color = 'green', icon = 'hand-up')).add_to(map_folium)
folium.Marker(location=[37.579804038387785, 126.97697662488285], popup=folium.Popup('경북궁', max_width = 200), icon = folium.Icon(color = 'blue')).add_to(map_folium)
folium.Marker(location=[37.577010787445786, 126.97690265925215], popup=folium.Popup('광화문', max_width = 200), icon = folium.Icon(color = 'black')).add_to(map_folium)
folium.Marker(location=[37.57072227958591, 126.98225914764045], popup=folium.Popup('영풍문고', max_width = 200), icon = folium.Icon(color = 'white')).add_to(map_folium)

map_folium.save('./Output/map.html')
map_folium
```
