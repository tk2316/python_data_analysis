판다스 데이터프레임과 시리즈    
데이터프레임과 시리즈는 리스트, 딕셔너리와 비슷하지만 데이터를 다루는데 더 특화되어 있다.   
리스트와 딕셔너리는 많은 양의 데이터를 저장하거나 조작할 수 있는 함수가 많치 않은 반면에 판다스의 데이터프레임과 시리즈는 많은 양을 저장할 수 있을 뿐만 아니라 스프레드시트 프로그램을 사용하는 것 처럼 행과 열 단위로 데이터를 조작할 수 있는 다양한 속성과 함수를 제공한다.

데이터프레임 만들기

```python
#데이터프레임을 만들기 위해서는 딕셔너리를 DataFrame() 함수의 인수로 전달하면 된다.
df = pd.DataFrame({
    # 'key': [val, val, ...] #key는 데이터프레임의 열 이름이 되고 value가 데이터가 된다.
    'Name' : ["홍길동","김태훈"],
    'Job' : ['의사', '조종사'],
    'Born' : ['1997-07-09', '1920-12-24'],
    'Died' : ['2012-01-12', '1945-10-12'],
    'Age' : [37, 60]
})
print(type(df))
df
```

```python
# DataFrame() 함수의 data 속성으로 데이터를 지정하고 index 속성으로 인덱스를 columns 속성으로 열 이름을 지정할 수 있다.
df = pd.DataFrame(
    data={
        'Job' : ['의사', '조종사'],
        'Born' : ['1997-07-09', '1920-12-24'],
        'Died' : ['2012-01-12', '1945-10-12'],
        'Age' : [37, 60]
    },
    index = ['홍길동','김태훈'],
    columns = ['Job','Age', 'Born', 'Died' ]     #columns를 생략하면 data에 쓴 순서데로 나온다. 그리고 data없는거를 column에 쓰면 NaN이라고 나온다. columns 이용해 나오는 순서를 바꿀수있다.
    #columns를 쓰면 columns 순서데로 나온다
)
print(type(df))
df

#간단한 데이터프레임은 만들 수 있는데 보통 데이터는 외부에서 다운받는다. 굳이 만들고 싶다면 excel을 이용하고 데이터를 입력하고나서 파일을 csv, tsv로 저장하면 된다.
```

index, values 속성과 keys() 함수와   

```python
#index 속성은 시리즈의 인덱스를 얻어온다.
print(type(row.index))
print(row.index)

print('*' * 80)
#values 속성은 시리즈의 데이터를 얻어온다
print(type(row.values))
print(row.values)

print('*' * 80)
# keys() 함수는 index 속성과 같이 시리즈의 인덱스를 얻어온다.
print(type(row.keys()))
print(row.keys())    #print(row.index)랑 결과가 똑같다

print('*' * 80)
#index 속성의 첫 번째 값 추출하기
print(row.index[0])
print(row.keys()[0])

print('*' * 80)
#values 속성의 첫 번째 값 출력하기
print(row.values[0])
```

```python
scientist_df = pd.read_csv('./Data/scientists.csv')
print(type(scientist_df))
scientist_df
```

```python
ages = scientist_df['Age']
print(type(ages))
ages

print('max = {}'.format(ages.max()))
print('min = {}'.format(ages.min()))
print('sum = {}'.format(ages.sum()))
print('mean = {}'.format(ages.mean()))
print('count = {}'.format(ages.count()))
print('std = {}'.format(ages.std()))
print('median = {}'.format(ages.median()))
```

브로드케스팅    
ages(나이 전체, 여러개) > ages.mean()(나이 평균, 1개)와 같이 시리즈나 데이터프레임의 모든 데이터에 대해 한 번에 연산하는 것을 브로드캐스팅이라 한다.   
이 떄, 여러개의 값을 가진 데이터를 백터라 하고 단순 크기를 나타내는 데이터를 스캘라라 부른다.

```python
# 평균 나이보다 나이가 많은 사람의 나이 데이터를 추출한다.
print(ages > ages.mean())    #브로드케스팅
bool_value = ages > ages.mean()
bool_value
```

```python
print(ages[bool_value])   #boolean indexing
```

```python
print(ages[ages > ages.mean()])     #boolean indexing
```

```python
for age in ages:
    print("{0:4d}".format(age), end=' ')
    
print()
print()
#같은 길이의 백터로 연산을 하면 결과값으로 같은 길이의 백터가 출력된다
for age in ages + ages:
    print("{0:4d}".format(age), end=' ')

    
print()
print()
for age in ages * ages:
    print("{0:4d}".format(age), end=' ')

#백터의 스칼라 값을 연산하면 백터 각각의 값에 스칼라 값이 연산된 결과가 출력된다
print()
print()
for age in ages + 100:
    print("{0:4d}".format(age), end=' ')

    
print()
print()
for age in ages * 2:
    print("{0:4d}".format(age), end=' ')
```

```python
print(pd.Series([1, 100]))
# 길이가 서로 다른 백터를 연산하면 같은 인덱스를 가지는 값만 계산된다.
# pd.Series([1, 100])의 0, 1 인덱스만 계산되고 나머지는 계산할 수 없으므로 누락값(NaN, 결측치)으로 처리한다.
print(ages + pd.Series([1, 100]))
```

```python
print(ages)
# sort_index() 함수는 데이터를 정렬하는 것이 아니고 인덱스를 정렬한다.
#ascending 옵션을 False로 지정하면 내림차순으로 정렬된다. 생략시 기본값은 True가 사용된다
#rev_ages = ages.sort_index(ascending=True)
#rev_ages = ages.sort_index()

rev_ages = ages.sort_index(ascending=False)
print(rev_ages)
#데이터를 정렬하려면 sort_values
rev_ages_val = ages.sort_values()
print(rev_ages_val)
rev_ages_val = ages.sort_values(ascending=False)
print(rev_ages_val)
```

```python
print(ages)
rev_ages = ages.sort_index(ascending=False)
print(rev_ages)
#백터와 백터의 연산은 같은 인덱스를 가지는 갋끼리 수행하기 때문에 ages + rev_ages와 ages * 2는 같은 결과가 계간된다
for age in ages + rev_ages:
    print("{0:4d}".format(age), end=' ')
print()
#0번 인덱스는 0번 인덱스랑 더한다
```

```python
print(ages)
rev_ages = ages.sort_values(ascending=False)
print(rev_ages)
#백터와 백터의 연산은 같은 인덱스를 가지는 갋끼리 수행하기 때문에 
for age in ages + rev_ages:
    print("{0:4d}".format(age), end=' ')
print()
#0번 인덱스는 0번 인덱스랑 더한다

for age in ages * 2:
    print("{0:4d}".format(age), end=' ')
print()
```

파이썬 strftime() 함수로 날짜 서식을 지정할 때 사용하는 날짜 서식    

%Y = 년도 4자리, %y = 년도 2자리,   
%m = 월(숫자), %B = 월(영어, 단어 전체), %b 아니면 %h (영어, 3글짜),    
%d = 일    
%A = 요일(전체), %a = 요일(3글짜)
%H = 시간(24시각), %I = 시간(12시각), %M = 분, %S = 초, %p = AM/PM, %f = 마이크로초(점 뒤에 6자리)


```python
#날짜를 문자열 형태로 저장한 데이터는 날짜 및 시간관련 작업을 할 수 있도록 to_datetime() 함수로 datetime 자료형으로 
# 바꿔주는 것이 좋다
#to_datetime(시리즈, format='%Y-%m-%d'), format 속성 생략시 '%Y-%m-%d'가 기본 서식으로 사용된다.

born_datetime = pd.to_datetime(scientist_df["Born"], format='%Y-%m-%d')
#born_datetime = pd.to_datetime(scientist_df["Born"], format='%y-%m-%d')
#ValueError: time data '1920-07-25' does not match format '%y-%m-%d' (match)
print(born_datetime)

died_datetime = pd.to_datetime(scientist_df["Died"])   # 생략시 자동으로 서식 format='%Y-%m-%d'으로 만들어진다
print(died_datetime)
```
