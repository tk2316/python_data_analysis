```python
# concat() 함수는 연결하려는 데이터프레임을 리스트에 담아서 전달하면 연결된 데이터프레임으로 반환한다.
# concat([데이터프레임1, 데이터프레임2, 데이터프레임3,...])
# 데이터프레임을 연결할 때 위에서 아래 방향으로 연결되고 df1, df2, df3 열 이름이 모두 A, B, C, D로 같고 데이터프레임을 
# 연결한 다음에도 열이 그대로 유지되고 인덱스도 그대로 유지된다.

row_concat = pd.concat([df1,df2, df3])
row_concat

```

```python
#데이터프레임에 시리즈 연결하기 => 반드기 데이터프레임으로 만들어서 연결해야 한다.
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])  # 리스트를 시리즈로 변환한다.
new_row_series
```

```python
pd.concat([df1, new_row_series])
```

```python
pd.DataFrame(['n1','n2','n3','n4'])   #열 단위의 데이터프레임이 생성된다
```

```python
# 열 이름을 지정하지 않으면 판다스가 0, 1, 2, 3과 같은 방식으로 열 이름을 붙어주기 때문에 columns 속성을 사용해서 열 이름을
# 지정해야 한다.

new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns = ['A', 'B', "C","D"])
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

```python
# concat() 함수는 한 번에 2개 이상의 데이터프레임을 연결하는 함수이다. 만약 연결할 데이터프레임 1개라면 append() 함수를 
# 사용해도 된다.
# 데이터프레임.append(추가할 데이터프레임)
df2.append(new_row_df)
```

```python
# concat(), append() 함수를 사용해서 데이터프레임을 연결할 떄 인덱스를 다시 부여히라면 ignore_index 옵션을 사용한다.
# 생략시 시본값은 False 이고 기존의 인덱스가 유지되지만 True로 지정하면 기존 인덱스가 무시되고 다시 부여된다.
df2.append(new_row_df, ignore_index=True)  #기본값은 False다
```

```python
#### append()함수와 딕셔너리를 사용하면 추가할 데이터를 판다스 데이터프레임으로 구성하지 않고 데이터프레임에 추가할 수 있다.
# 데이터프레임.append(추가할 데이터가 저장된 딕셔너리)
data_dict = {'A' :'n5', 'B' : 'n6', 'C' : 'n7', 'D' : 'n8'}
print(data_dict)
df3.append(data_dict, ignore_index=True)
```

```python
# concat() 함수는 axis 속성을 생략하면 0이 기본값으로 사용되며 아래쪽으로 데이터프레임이 추가된다
row_concat = pd.concat([df1,df2, df3], ignore_index=True, axis = 0) 
row_concat
```

```python
col_concat = pd.concat([df1,df2, df3], ignore_index=True, axis = 1) #axis 1이면 열방향으로 (우측 방향으로) 데이터프레임이 추가된다.
col_concat
#거의 안쓴다. 보통 믿으로 데이터는 추가된다.
```

```python
person = pd.read_csv('./Data/survey_person.csv')
person
```

```python
site = pd.read_csv('./Data/survey_site.csv')
site
```

```python
survey = pd.read_csv('./Data/survey_survey.csv')
survey
```

```python
visited = pd.read_csv('./Data/survey_visited.csv')
visited
```

```python
#지금 4개의 다른 데이터프레임이 있는데 이제 합쳐서 하나의 데이터프레임을 만드는 작업할것이다
#merge() 함수는 기본적으로 내부(inner) 조인을 실행하며 함수를 실행한 데이터프레임이 왼쪽(left) 으로 지정되고 첫 번째 인수로 지정된
#데이터프레임이 오른(right)쪽으로 지정되며 left_on 속성과 right_on 속성은 각각 왼쪽과 오른쪽 데이터프레임에서 일치해야
# 할 열을 지정한다
print(site.columns)
print(visited.columns)
o2o_merge = site.merge(visited, left_on = 'name', right_on = 'site')     #site왼쪽 visited은 오른쪽
o2o_merge
```

```python
print(person.columns)
print(survey.columns)
p2s_merge = person.merge(survey, left_on = 'ident', right_on = 'person')
p2s_merge
```

```python
print(visited.columns)
print(survey.columns)
v2s_merge = visited.merge(survey, left_on='ident', right_on='taken')
v2s_merge
```

```python
print(p2s_merge.columns)
print(v2s_merge.columns)

# 중복된 열 이름에는 _x, _y가 추가된다. _x 왼쪽 데이터프레임을 _y는 오른쪽 데이터프레임을 의미한다.
p2s_v2s_merge = p2s_merge.merge(v2s_merge, left_on= ['ident','taken','quant','reading'], right_on = ['person','ident','quant','reading'])
p2s_v2s_merge
```
