melt()함수
데이터프레임을 깔끔한 형태로 정리하는 데 사용하는 melt() 함수는 지정한 열의 데이터를 모두 행으로 정리한다.   
melt(데이터프레임, id_vars[, value_vars, var_name, value_name])

id_vars: 위치를 그대로 유지할 열 이름    
value_vars: 행으로 위치를 변경할 열 이름, 생략시 모든 열을 행으로 변경한다.   
var_name: value_vars로 위치를 변경한 열 이름을 지정한다. 생략시 variable로 작  성된다.    
value_name: var_name으로 위치를 변경한 열의 데이터에 지정할 열 이름을 지정한다. 생략시 value로 작성된다.

```python
# 종교(religion) 열은 위치를 그대로 유지하고 열을 구성하고 있는 소득 정보 데이터를 행 데이터로 옮긴다
# 이런 과정을 'religion' 열을 고정해서 '피벅했다' 라고 말한다.
pew_long = pd.melt(pew, id_vars='religion')
pew_long
```

```python
# value_vars 속성을 지정하면 지정한 열만 대상으로 피벗한다.
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k') #피벗할 열이1개
pew_long
```

```python
#피벗할 열이 2개 이상이라면 피벗할 열을 리스트로 묶어 value_vars 속성에 전달하면 된다.
pew_long = pd.melt(pew, id_vars ='religion', value_vars=['<$10k','$10-20k'])  #피벗할 열이 1개 이상
pew_long
```

```python
# var_name 속성을 이용해서  variable로 표시되는 열 이름을 변경한다.
# value_name 소성을 이용해서 value로 표시되던 열 이름을 변경한다
pew_long = pd.melt(pew, id_vars ='religion', value_vars='$10-20k', var_name = '최저임금', value_name='인원수')
pew_long
```

```python
# data와 day를 고정하고 나머지를 행으로 피벗하면 각 나라별 환자수와 사망자 수를 볼 수 있다
ebola_long = pd.melt(ebola, id_vars = ['Date', 'Day'], var_name = "Country", value_name = 'Cases')
ebola_long
```

```python
# data와 day를 고정하고 나머지를 행으로 피벗하면 각 나라별 환자수와 사망자 수를 볼 수 있다
ebola_long = pd.melt(ebola, id_vars = ['Date', 'Day'])
ebola_long.columns  = ['날짜', '경과일', '국가명', '확진자']
ebola_long
```

```python
# split() 함수의 인수에  '_'를 구분자로 전달하면 Cases_Guinea를 Cases와 Guinea로 분리할 수 있다.
print(ebola_long['국가명'])
print()
variable_str = ebola_long.국가명.str    #Series는 문자열 함수를 사용할 수 없으므로 str을 사용해서 문자열로 변환한 후 사용한다.
variable_split = variable_str.split('_')
```

```python
print(variable_split)    
print(type(variable_split)) #series
print(type(variable_split[0]))    #list
print(variable_split[0][0])    #string
print(variable_split[0][1])
```

```python
status_values = variable_split.str[0]
print(status_values.head())
country_values = variable_split.str[1]
print(country_values.head())
type(status_values)
```

```python
ebola_long['상태'] = status_values
ebola_long['국가'] = country_values
ebola_long
```

```python
variable_split = ebola_long.국가명.str.split("_")
print(type(variable_split))
# split() 함수에 expand 옵션을 True로 지정하면 구분자를 경계로 분리한 데이터를 데이터프레임으로 리턴한다.
variable_split = ebola_long.국가명.str.split("_", expand=True)   #제일 쉬운 방법 => 바로 데이터프레임으로 만든다
print(type(variable_split))
variable_split.head()

```
